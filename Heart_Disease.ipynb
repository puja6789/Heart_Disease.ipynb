# Task
Write a basic machine learning project using a suitable dataset. The project should include data loading, exploration, preprocessing, model training, and evaluation.

## Load data

### Subtask:
Load a dataset suitable for a basic machine learning task (e.g., a classification or regression problem).


**Reasoning**:
The first step is to import the pandas library to work with dataframes and then load a suitable dataset for a machine learning task. The Iris dataset from sklearn.datasets is a good choice for a basic classification problem and is easily accessible.



```python
import pandas as pd
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['target'] = iris.target

# Display the first few rows of the dataframe
print("First 5 rows of the dataset:")
display(df.head())

# Display information about the dataset
print("\nDataset Info:")
df.info()

# Display basic statistics
print("\nBasic Statistics:")
display(df.describe())
```

## Explore data

### Subtask:
Perform basic exploratory data analysis to understand the data structure, types, and distributions.


**Reasoning**:
Perform exploratory data analysis by checking data types, missing values, descriptive statistics, and target distribution as per the instructions.



```python
# 1. Display column names and their data types
print("Column names and data types:")
print(df.dtypes)

# 2. Check for missing values
print("\nMissing values per column:")
print(df.isnull().sum())

# 3. Display descriptive statistics for numerical columns
print("\nDescriptive statistics for numerical columns:")
display(df.describe())

# 4. Determine the number of unique values in the 'target' column and their distribution
print("\nNumber of unique values in 'target' column:", df['target'].nunique())
print("\nDistribution of unique values in 'target' column:")
print(df['target'].value_counts())
```

## Preprocess data

### Subtask:
Prepare the data for machine learning by handling missing values, encoding categorical features, and scaling numerical features.


**Reasoning**:
Prepare the data for machine learning by separating features and target, scaling features, and splitting the data into training and testing sets.



```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Separate features (X) and target (y)
X = df.drop('target', axis=1)
y = df['target']

# Scale the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

print("Shape of X_train:", X_train.shape)
print("Shape of X_test:", X_test.shape)
print("Shape of y_train:", y_train.shape)
print("Shape of y_test:", y_test.shape)
```

## Train model

### Subtask:
Select and train a simple machine learning model on the preprocessed data.


**Reasoning**:
Import the LogisticRegression model and train it using the training data.



```python
from sklearn.linear_model import LogisticRegression

# Instantiate the Logistic Regression model with default parameters
model = LogisticRegression()

# Train the model using the training data
model.fit(X_train, y_train)

print("Model training complete.")
```

## Evaluate model

### Subtask:
Evaluate the trained model using appropriate metrics to assess its performance.


**Reasoning**:
Evaluate the trained model using appropriate metrics.



```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Make predictions on the test set
y_pred = model.predict(X_test)

# Calculate evaluation metrics
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')
f1 = f1_score(y_test, y_pred, average='weighted')

# Print the evaluation metrics
print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1-score: {f1:.4f}")
```

## Summary:

### Data Analysis Key Findings

*   The Iris dataset, containing 150 instances and 5 columns (4 features and 1 target), was successfully loaded.
*   The dataset has no missing values, and the target variable has three unique classes (0, 1, 2) with an equal distribution of 50 instances per class.
*   The features were successfully scaled using `StandardScaler` and the dataset was split into training and testing sets with an 80/20 ratio.
*   A Logistic Regression model was trained on the scaled training data.
*   The trained model achieved perfect scores (1.0000) for Accuracy, Precision, Recall, and F1-score on the test set.

### Insights or Next Steps

*   The model's perfect performance on this dataset suggests it is a relatively easy classification problem for Logistic Regression, or potentially indicates data leakage or overfitting if not using a robust validation strategy.
*   For a more comprehensive evaluation and to build a more robust model, consider implementing cross-validation and exploring other classification algorithms.


# Task
Write a basic machine learning project for heart disease detection.

## Load data

### Subtask:
Load a dataset containing medical information that can be used to predict heart disease.


**Reasoning**:
Import the pandas library and load the heart disease dataset from the specified URL into a pandas DataFrame. Display the first few rows and the shape of the DataFrame to inspect the loaded data.



```python
import pandas as pd

# Load the dataset
# Using a publicly available dataset from UCI Machine Learning Repository
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/heart-disease/processed.cleveland.data"
# Define column names as per the dataset documentation
column_names = [
    "age", "sex", "cp", "trestbps", "chol", "fbs", "restecg",
    "thalach", "exang", "oldpeak", "slope", "ca", "thal", "target"
]
df = pd.read_csv(url, names=column_names, na_values='?')

# Display the first few rows
print("First 5 rows of the dataset:")
display(df.head())

# Print the shape of the DataFrame
print("\nShape of the dataset:", df.shape)
```

## Explore data

### Subtask:
Perform exploratory data analysis to understand the data, including checking for missing values, visualizing distributions, and identifying potential correlations.


**Reasoning**:
Perform exploratory data analysis as per the instructions, including checking data types, missing values, descriptive statistics, correlations, and target distribution.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Check the data types of each column
print("Data types of each column:")
print(df.dtypes)

# 2. Check for and count any missing values
print("\nMissing values per column:")
print(df.isnull().sum())

# 3. Display descriptive statistics for the numerical columns
print("\nDescriptive statistics for numerical columns:")
display(df.describe())

# 4. Calculate and display the correlation matrix
print("\nCorrelation Matrix:")
plt.figure(figsize=(12, 10))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Features')
plt.show()

# 5. Display the distribution of the 'target' column
print("\nDistribution of the 'target' column:")
print(df['target'].value_counts())

plt.figure(figsize=(6, 4))
sns.countplot(x='target', data=df, palette='viridis')
plt.title('Distribution of Target Variable')
plt.xlabel('Target')
plt.ylabel('Count')
plt.show()
```

## Preprocess data

### Subtask:
Prepare the data for machine learning, which may include handling missing values, encoding categorical features, scaling numerical features, and splitting the data into training and testing sets.


**Reasoning**:
Prepare the data for machine learning by handling missing values, converting the target to binary, separating features and target, scaling features, and splitting the data into training and testing sets.



```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# 1. Handle missing values in 'ca' and 'thal' by filling with the mode
for col in ['ca', 'thal']:
    mode_value = df[col].mode()[0]
    df[col].fillna(mode_value, inplace=True)

# Check if missing values are handled
print("Missing values after filling with mode:")
print(df.isnull().sum())

# 2. Convert the 'target' column into a binary format (0 for no heart disease, 1 for heart disease)
df['target'] = df['target'].apply(lambda x: 1 if x > 0 else 0)

print("\nDistribution of binary target column:")
print(df['target'].value_counts())

# 3. Separate features (X) and target (y)
X = df.drop('target', axis=1)
y = df['target']

# 4. Scale the numerical features
# Identify numerical columns (all are numerical in this dataset)
numerical_cols = X.columns

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X[numerical_cols])

# Convert scaled features back to a DataFrame (optional, but good practice)
X_scaled_df = pd.DataFrame(X_scaled, columns=numerical_cols, index=X.index)

# 5. Split the scaled data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled_df, y, test_size=0.2, random_state=42, stratify=y)

print("\nShape of X_train:", X_train.shape)
print("Shape of X_test:", X_test.shape)
print("Shape of y_train:", y_train.shape)
print("Shape of y_test:", y_test.shape)
```

## Train model

### Subtask:
Select and train a suitable machine learning model for classification (since heart disease detection is a classification problem).


**Reasoning**:
Import the LogisticRegression model and train it using the training data.



```python
from sklearn.linear_model import LogisticRegression

# Instantiate the Logistic Regression model with default parameters
model = LogisticRegression()

# Train the model using the training data
model.fit(X_train, y_train)

print("Model training complete.")
```

## Evaluate model

### Subtask:
Evaluate the trained model using appropriate classification metrics such as accuracy, precision, recall, F1-score, and AUC-ROC.


**Reasoning**:
Evaluate the trained model using appropriate classification metrics as per the instructions.



```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

# Make predictions on the test set
y_pred = model.predict(X_test)

# Calculate evaluation metrics
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')
f1 = f1_score(y_test, y_pred, average='weighted')

# Calculate AUC-ROC score for binary classification
# Need predicted probabilities for roc_auc_score
y_pred_proba = model.predict_proba(X_test)[:, 1]
auc_roc = roc_auc_score(y_test, y_pred_proba)

# Print the evaluation metrics
print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1-score: {f1:.4f}")
print(f"AUC-ROC: {auc_roc:.4f}")
```

## Summary:

### Data Analysis Key Findings

*   The dataset contains 303 entries and 14 columns.
*   Columns 'ca' and 'thal' had missing values (4 and 2 respectively), which were later imputed with the mode.
*   The target variable distribution was imbalanced before preprocessing, with 164 instances of no heart disease and 139 instances of heart disease after converting to a binary format.
*   After training a Logistic Regression model, the evaluation metrics on the test set were:
    *   Accuracy: 86.89%
    *   Precision: 87.66%
    *   Recall: 86.89%
    *   F1-score: 86.90%
    *   AUC-ROC: 95.13%

### Insights or Next Steps

*   The high AUC-ROC score suggests the model has good capability to distinguish between patients with and without heart disease.
*   Further steps could involve exploring other classification models, hyperparameter tuning, or feature engineering to potentially improve performance, although the current results are promising.
